```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
creditcard_data=pd.read_csv('creditcard.csv')
creditcard_data.head()
```

```python
creditcard_data.tail()
```

```python
creditcard_data.info()
```

```python
creditcard_data.isnull().sum()
```

```python
creditcard_data['Class'].value_counts()
```

```python
legit=creditcard_data[creditcard_data.Class==0]
```

```python
fraud=creditcard_data[creditcard_data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
creditcard_data
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
creditcard_data.groupby('Class').mean()
```

```python
legit_sample=legit.sample(n=492)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
```

```python
Y=new_dataset['Class']
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

ACCURACY ON TRAINING DATA

```python
X_train_prediction=model.predict(X_train)
traning_data_accuracy=accuracy_score(X_train_prediction,Y_train)
traning_data_accuracy
```

ACCURACY ON TRAINING DATA

```python
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
testing_data_accuracy
```
